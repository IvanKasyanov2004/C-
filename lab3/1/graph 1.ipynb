```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

N = []
T = []
with open('1.txt', 'r') as file:
    for line in file:
        N.append(float(line.split()[0]))
        T.append(float(line.split()[1]) / 1000)

plt.scatter(N, T, s = 40, color = (0,0,1))

plt.title(r"График 1. T(N) (1000 запусков)")
plt.xlabel(r"N")
plt.ylabel(r"T, c")
plt.grid(True)
plt.minorticks_on()
plt.grid(which = 'minor', alpha = 0.2)

n = np.linspace(0, 1000, 100)
t = np.array(4*10**-6 * n ** 2)
plt.plot(n, t, color = (1,0,0), linewidth = 2, label = r"$T = 4 \cdot 10^{-6}$ $N^2$")
plt.legend()
```

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

N = []
T = []
with open('1_steps.txt', 'r') as file:
    for line in file:
        N.append(float(line.split()[0]))
        T.append(float(line.split()[1]) / 10**6)

plt.scatter(N, T, s = 40, color = (0,0,1))

plt.title(r"График 2. steps(N) (1000 запусков)")
plt.xlabel(r"N")
plt.ylabel(r"$steps \cdot 10^{-6}$")
plt.grid(True)
plt.minorticks_on()
plt.grid(which = 'minor', alpha = 0.2)

n = np.linspace(0, 1000, 100)
t = np.array(0.14 / 10**3 * n ** 2)
plt.plot(n, t, color = (1,0,0), linewidth = 2, label = r"$T = 140$ $N^2$")
plt.legend()
```
